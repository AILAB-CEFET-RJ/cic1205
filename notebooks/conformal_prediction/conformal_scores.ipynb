The following code illustrates the computation of the conformal threshold using a small set of calibration scores. The example follows the same three steps described in the lecture notes: sorting the scores, computing the corrected rank, and selecting the corresponding ordered score as $\hat{q}$. It also highlights the indexing difference between the mathematical notation, where $s_{(r)}$ starts at position 1, and Python indexing, where arrays start at position 0.

`````python
import numpy as np
import math

# Calibration scores: for example, absolute errors on the calibration set
calibration_scores = np.array([
    0.12, 0.35, 0.08, 0.21, 0.44,
    0.18, 0.27, 0.31, 0.05, 0.39
])

# Target miscoverage level
alpha = 0.10  # target coverage = 1 - alpha = 90%

# Number of calibration examples
n_cal = len(calibration_scores)

# Step 1: sort scores in increasing order
sorted_scores = np.sort(calibration_scores)

# Step 2: compute the corrected rank
r = math.ceil((n_cal + 1) * (1 - alpha))

# Step 3: obtain the conformal threshold
# Python uses 0-based indexing, so s_(r) is stored at index r - 1
if r <= n_cal:
    q_hat = sorted_scores[r - 1]
else:
    q_hat = np.inf

print("Calibration scores:")
print(calibration_scores)

print("\nSorted calibration scores:")
print(sorted_scores)

print(f"\nNumber of calibration examples: n_cal = {n_cal}")
print(f"Target coverage: 1 - alpha = {1 - alpha:.2f}")
print(f"Corrected rank: r = ceil(({n_cal} + 1) * {1 - alpha:.2f}) = {r}")

print(f"\nConformal threshold: q_hat = {q_hat}")
`````

When the corrected rank $r$ is larger than the number of calibration examples $(n_{\text{cal}}$), there is no calibration score $s_{(r)}$ to select.

In that case, the code sets

```python
q_hat = np.inf
````

This means that the conformal threshold is treated as infinity. As a consequence, the resulting prediction interval becomes infinitely wide:

$
[\hat{y} - \infty,\ \hat{y} + \infty].
$

Although this interval is not useful in practice, it is a mathematically safe convention: it preserves the desired coverage guarantee when the calibration set is too small, or when the requested confidence level is too high for the available number of calibration scores.

In practical applications, this situation is a warning sign. It indicates that we may need more calibration data, or that the chosen value of $\alpha$ leads to a corrected rank beyond the available calibration scores.
